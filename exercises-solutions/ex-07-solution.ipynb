# Problem

Write a function `calc_chi2` that calculates the reduced chi-squared value between a model and a data set:

$$
\chi^2 \,=\, \frac{1}{N}\,\sum_{i=0}^{N}\frac{(F_i-F^\rm{model}_i)^2}{\sigma_i^2},
$$

where $N$ is the number of points in your data set, and $F^\rm{model}_i$ is the flux value predicted by the model for energy $E_i$.

Then, use your function to get the $\chi^2$ value between the **model given by the magenta curve** and the **data measured at energies above 2 eV**.

```python
def calc_chi2(modx, mody, datx, daty, daterr):
    """Calculate the reduced chi^2 between data and model curve.
    
    Args:
        modx (np.array): energy array of the model curve
        mody (np.array): flux array of the model curve
        datx (np.array): data energy values (same units as modx)
        daty (np.array): data flux values (same units as mody)
        daterr (np.array): data uncertainties (same units as daty)
    Returns:
        redchi2 (float): reduced chi-squared value
    """

    ainterp = np.interp(datx, modx, mody)
    avarian = (ainterp - daty) ** 2 / daterr ** 2
    chi2val = np.sum(avarian)
    redchi2 = chi2val / datx.size
    return redchi2


irange = np.where(pks_data_x > 2.)    # Select data above 2 eV
chi2 = calc_chi2(pks_model_x,         # Calculate the chi squared
                 pks_model_y,
                 pks_data_x[irange],
                 pks_data_y[irange],
                 pks_data_err[irange])

print(f"Reduced chi2 = {chi2:.3f}")
```
